## Scrape and Analyze Player Data from fbref.com website
Create a Beautiful Soup object and use it to scrape the data in the HTML table


```python
# importing required libraries

import pandas as pd
import numpy as np
from bs4 import BeautifulSoup as soup
from splinter import Browser
```

```python
browser = Browser('chrome')
```

```python
#urls for Player Stats seasons 2017_2019 to 2021_2022

url_1718_player = "https://fbref.com/en/comps/8/2017-2018/stats/2017-2018-Champions-League-Stats"
url_1819_player = "https://fbref.com/en/comps/8/2018-2019/stats/2018-2019-Champions-League-Stats"
url_1920_player = "https://fbref.com/en/comps/8/2019-2020/stats/2019-2020-Champions-League-Stats"
url_2021_player = "https://fbref.com/en/comps/8/2020-2021/stats/2020-2021-Champions-League-Stats"
url_2122_player = "https://fbref.com/en/comps/8/2021-2022/stats/2021-2022-Champions-League-Stats"  
```

```python
# Visitng the urls and parsing the html using BeautifulSoup

urls = [url_1718_player, url_1819_player, url_1920_player, url_2021_player, url_2122_player]
data = []

for url in urls:
    browser.visit(url)
    browser.is_element_present_by_css('div.list_text', wait_time=1)
    html = browser.html
    player_soup = soup(html, 'html.parser')
    
    #selecting the player's table
    
    table = player_soup.find('table', class_='min_width sortable stats_table shade_zero now_sortable sticky_table eq2 re2 le2')
    
    columns = []
    col_rows = table.find('thead').find_all('tr')    
    header_row = col_rows[1] # table had 2 header rows and we wanted the second header row only for column names 
    for header in header_row.find_all('th'):
        columns.append(header.text)   
    
    print(columns)    
    
    # Create an empty list
    table_data = []
    
    # Loop through the scraped data to create a list of rows

    rows = table.find_all('tr')

    for row in rows[2:]:   # first 2 rows are column header
        values = []
        row_data = row.find_all('td')
        for i in row_data:
            values.append(i.text)
        table_data.append(values)
   
    data.append(table_data)
    
    print(data)
```

```python
print(columns[1:])
```

## Creating Dataframes from the parsed html data

```python
player_17_18_df = pd.DataFrame(data[0], columns = columns[1:])
player_17_18_df['Seasons']= '2017-2018'
player_17_18_df.head()
```

```python
player_18_19_df = pd.DataFrame(data[1], columns = columns[1:])
player_18_19_df['Seasons']= '2018-2019'
player_18_19_df.tail()
```

```python
player_19_20_df = pd.DataFrame(data[2], columns = columns[1:])
player_19_20_df['Seasons']= '2019-2020'
player_19_20_df.tail()
```

```python
player_20_21_df = pd.DataFrame(data[3], columns = columns[1:])
player_20_21_df['Seasons']= '2020-2021'
player_20_21_df.tail()
```

```python
player_21_22_df = pd.DataFrame(data[4], columns = columns[1:])
player_21_22_df['Seasons']= '2021-2022'
player_21_22_df.tail()
```

```python
player_all_seasons_df = pd.concat([player_17_18_df, player_18_19_df, player_19_20_df, player_20_21_df, player_21_22_df])
player_all_seasons_df
```

```python
#some columns have duplicate names. Selecting required columns
final_player_df = player_all_seasons_df.iloc[:-1,[0,3,6,7,8,9,10, -1]]
final_player_df
```

```python
final_player_df = final_player_df.rename(columns={'Gls':'Goals'})
final_player_df.head()
```

```python
final_player_df['League'] = 'UEFA'
final_player_df.head()
```

```python
final_player_df.to_csv('Data/player_UEFA)',  index=False)
```
